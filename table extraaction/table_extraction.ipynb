The 403 Forbidden error happens because Wikipedia (and many other sites) identifies the default "User-Agent" of Python's urllib as a bot and blocks the request to prevent scraping.

To get around this, you need to provide a User-Agent header so the request looks like it’s coming from a standard web browser.

```python

from io import StringIO

url = 'https://en.wikipedia.org/wiki/List_of_The_Simpsons_episodes_(seasons_1%E2%80%9320)'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)

# Wrap the text in StringIO to resolve the FutureWarning
s_list = pd.read_html(StringIO(response.text))

print(s_list[0].head())
```
